```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(size = 1000)
```

```python
plt.hist(x);
plt.show()
```

```python
plt.hist(x, normed = True);
plt.show()
```

```python
plt.hist(x, normed = True, bins = np.linspace(-5, 5, 25));
plt.show()
```

```python
plt.hist(x, normed = True, bins = np.linspace(-5, 5, 225));
plt.show()
```

```python
x = np.random.gamma(2,3,100000)
```

```python
plt.hist(x, bins = 30)
plt.show()
```

```python
plt.hist(x, bins = 30, normed = True)
plt.show()
```

```python
plt.hist(x, bins = 30, cumulative = True)
plt.show()
```

```python
plt.hist(x, bins = 30,normed = True, cumulative = True, histtype = 'step')
plt.show()
```

```python
plt.figure();
plt.subplot(221)
plt.hist(x, bins = 30)
plt.subplot(222)
plt.hist(x, bins = 30, normed = True)
plt.subplot(223)
plt.hist(x, bins = 30, cumulative = True)
plt.subplot(224)
plt.hist(x, bins = 30,normed = True, cumulative = True, histtype = 'step')
plt.show()
```
